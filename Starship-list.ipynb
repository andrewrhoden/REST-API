loop through all pages and store in a variable 

```python
url='https://swapi.dev/api/starships/?page='
pages=4
for i in range(1,pages+1):
    response=requests.get(url+str(i))
    fdata=response.json()
```

```python
pp(fdata)
```

retrieve the results dictionary with the values you need to manipulate


```python
data_results=[]
for i in range(0,4):
    data_results.extend(fdata['results'])
pp(data_results)

```

Write a function that returns all of the models of starships

```python
def models():
    ships_model=[]
    for i in data_results:
        ships_model.append(i['model'])
    return ships_model

pp(models())
        
        
```
